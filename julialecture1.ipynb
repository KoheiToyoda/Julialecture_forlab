## 1.2 How to install
プログラミングを行うためには、まずソースコード(命令書)を書き、  
そのソースコードを機械語に変換するためのコンパイラ、もしくはインタプリタというものが必要です。   
参考: https://www3.cuc.ac.jp/~miyata/classes/prg1/02/2way.html

Juliaはインタプリタ型ですが、  
処理のまとまっている部分でコンパイルをかけて実行するJIT（Just in time)コンパイラを採用しているため、  
インタプリタ型が持つ書きやすさ、実行のしやすさと  
コンパイラ型が持つ高速性を併せ持っています。  
  
まず、Juliaインタプリタを入手します。  
  
**Juliaインタプリタ**  
https://julialang.org/downloads/

**Windows10の人は、64-bit(installer)を選べばいい**と思います。

```C:\Users\$USERNAME$\AppData\Local\Programs\Julia x.x.x```  
x.x.xはバージョン番号です。  
このままで問題ないです。

## 1.3 REPLの使い方  
今インストールして立ち上がったJuliaは**REPL**というものです。  
REPLは**プログラムを読み込み(Read)、評価(Eval)、実行(Play)、を繰り返す(Loop)**、  
基本プログラムであり、このREPLを通して、プログラムの実行を可能です。  
REPLだけでプログラム実行をするのは、見通しが悪いので、  
REPLを操作するためのVSCodeやJupyterといった、統合開発環境（IDE)というものをあとからインストールします。  
  
しかし、JuliaのREPLには便利な機能があるので、まずはREPLの2大モードについて学習しましょう。  


### 1.3.1 HELPモード

```julia> ```  

と出てきたら、カーソルを合わせて、

``` ? ```  

を入力してください。  
``` (HELP)> ```  


これだけでjuliaはHELPモードに入ります。  
分からないことはここに入力すればいいです。  
たとえば   
  
``` for ```    
  
と入力してみましょう。 

  
```
for loops repeatedly evaluate a block of statements while iterating over a sequence of values.

  Examples
  ≡≡≡≡≡≡≡≡≡≡

  julia> for i in [1, 4, 0]
             println(i)
         end
  1
  4
  0
```

このように基本構文から、パッケージの使い方まで教えてくれます。  
  
HELPモードからはBackSpaceで抜けることができます。  

### 1.3.2 Pkg モード


``` Julia> ```  

に戻ったら

``` ] ```  

と入力してみてください  

``` (@v1.5)Pkg> ```  
これはPkg(パッケージ)モードというものです。  
Juliaは様々な機能を外部パッケ―ジとして提供しています。 
  
プロット機能はPlots.jl  
行列計算はLinearAlgebla.jl  
  
など、今なお現在、多くのパッケージが開発が進められています。  
これをダウンロードしてインストールしたり、  
削除したりするのがパッケージモードです。  
(pythonでいうpipやcondaのようなものです。)  
  

今回は以下のようにしてください。  
``` (@v1.5)Pkg> add Plots```   
これだけでJuliaにプロット機能を追加することができます。  
  
自動でダウンロードとインストールが始まります。  
終わったら次は以下のコマンドを打ちます。  
``` (@v1.5)Pkg> precompile```  
実際ソースコードを動かす前に事前にコンパイルしておくことで、  
プログラムを動かすのに必要な時間短縮が可能です。  
  
Pkgモードには他に、  
``` (@v1.5)Pkg> st ```    
``` (@v1.5)Pkg> status ```  
現在インストールされているパッケージ確認  
``` (@v1.5)Pkg> up ```    
``` (@v1.5)Pkg> update ```  
現在インストールされているパッケージを一括アップデート  
``` (@v1.5)Pkg> rm("****")```  
``` (@v1.5)Pkg> remove("****")```  
```**** ```という名前のパッケージを削除
等があります。
